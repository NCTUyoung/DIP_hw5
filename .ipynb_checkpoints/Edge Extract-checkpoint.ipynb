```python
%matplotlib inline
import PIL as PIL
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import cv2
from skimage import filters
plt.rcParams["figure.figsize"] = (12,8)
im_ori = np.array(PIL.Image.open('input1.jpg'))
im = np.array(PIL.Image.open('local_k.bmp').resize((im_ori.shape[1],im_ori.shape[0]),PIL.Image.ANTIALIAS))
plt.imshow(im)
plt.show()
im.shape
```

### Use image dilate/erode to get edge

```python
val = filters.threshold_otsu(im)
kernel = np.ones((3,3),np.uint8)
kernel2 = np.ones((4,4),np.uint8)

im_vgg = ((im>val)*1).astype(np.uint8)

output = cv2.dilate(im_vgg,kernel) - im_vgg
plt.imshow(output)
plt.show()
```

### Mask out edge area, turn it into red(255,0,0)

```python

im_or = im_ori.copy()
im_or[output>0,:] = [255,0,0]

im_out =  PIL.Image.fromarray(im_or).save('output_local.bmp')
plt.imshow(im_or)
plt.show()
```

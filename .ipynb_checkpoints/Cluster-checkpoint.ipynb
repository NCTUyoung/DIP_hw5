### Feature Vector Reform
use HSV feature

```python
feature_x = np.zeros((3,im[:,:,0].flatten().shape[0]))
x_index = np.linspace(0,1,im.shape[0])
y_index = np.linspace(0,1,im.shape[1])
grid = np.indices((im.shape[0], im.shape[1]))
im_hsv = color.rgb2hsv(im)

# feature_x[0,:] = grid[0].flatten()/(im.shape[0]-1)
# feature_x[1,:] = grid[1].flatten()/(im.shape[1]-1)
# feature_x[2,:] = im[:,:,0].flatten()/255.0
# feature_x[3,:] = im[:,:,1].flatten()/255.0
# feature_x[4,:] = im[:,:,2].flatten()/255.0
feature_x[0,:] = im_hsv[:,:,0].flatten()
feature_x[1,:] = im_hsv[:,:,1].flatten()
feature_x[2,:] = im_hsv[:,:,2].flatten()

sample = np.random.randint(0,len(feature_x[0,:]),size=2000)
# print(sample)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
c, m, zlow, zhigh = ('r', 'o', 0, 1)
xs = feature_x[0,sample]
ys = feature_x[1,sample]
zs = feature_x[2,sample]
ax.scatter(xs, ys, zs, c=c, marker=m)
ax.set_xlabel('R Label')
ax.set_ylabel('G Label')
ax.set_zlabel('B Label')
plt.show()
```

### Define K mean parameter

```python
K = 3
```

### Initial parameter

```python
mu =np.random.random_sample((3,K))
```

### K-means Algorthm

```python
r = np.zeros((K,im[:,:,0].flatten().shape[0]))
def calculate_r(a):
    r_vector = np.zeros(K)
    
    for i in range(K):
        r_vector[i] = np.square(a-mu[:,i]).sum()
#     np.argmax(r_vector)
    arg = np.zeros(K)
    arg[np.argmin(r_vector)] = 1
    return arg
r = np.apply_along_axis(calculate_r,0,feature_x)
```

### Update mu

```python
for i in range(5):
    acumulate_K = r.sum(axis=1)
    acumulate_mu = np.dot(r,feature_x.transpose())
    mu = np.divide(acumulate_mu,acumulate_K.reshape((K,1))).transpose()
    
    r = np.apply_along_axis(calculate_r,0,feature_x)
```

```python
out = np.argmax(r,axis=0).reshape(im[:,:,0].shape)
plt.imshow(out)
plt.show()
plt.imshow(im)
plt.show()
o = np.array(out==0)*1
o[o==1] = 255
# PIL.Image.fromarray(o).save('k_mean.bmp')
```

```python
import pandas as pd 
feature_x_pd = pd.DataFrame(data=feature_x.transpose(),columns=['H','S','V'])  # 1st row as the column names
feature_x_pd['class'] = pd.Series(np.argmax(r,axis=0))
feature_x_pd.head()
```

### EM Algorithm for Gaussian Mixture Models

### Alpha Cov init

```python
Alpha = np.random.random_sample(K)
Alpha

gmm_cov = np.zeros((K,feature_x.shape[0],feature_x.shape[0]))
data = feature_x_pd[['H','S','V']].groupby(feature_x_pd['class'])
for i in range(K):
    gmm_cov[i] = data.get_group(0).cov().values
gmm_cov
```

### MembershipWeights
![Alt text](./image/2017-11-30_202837.png)

```python
def calculate_w(x):
    w = np.zeros(K)
    for i in range(K):
        w[i] = multivariate_normal.pdf(x, mean=mu[:,i], cov=gmm_cov[i])*Alpha[i]
    return w/w.sum()
W = np.apply_along_axis(calculate_w,0,feature_x)
```

### Mstep

```python
Alpha = W.sum(axis=1)/W.sum(axis=1).sum()
N = W.sum(axis=1)
```

```python
W.shape
```

```python
mu_gmm = np.zeros((3,feature_x.shape[1])) 
mu_gmm = np.dot(W,feature_x.transpose())
mu_gmm = mu_gmm.transpose()
mu_gmm = mu_gmm/N
```

```python
mu_gmm[:,0,None].shape
gmm_cov.shape
```

```python
diff = feature_x-mu_gmm[:,0,None]

np.dot(np.multiply(diff,W[0]),diff.transpose())/N[0]
```

```python

for i in range(3):
    diff = feature_x-mu_gmm[:,i,None]
    gmm_cov[i] = np.dot(np.multiply(diff,W[i]),diff.transpose())/N[i]
```

```python
def calculate_w_new(x):
    w = np.zeros(K)
    for i in range(K):
        w[i] = multivariate_normal.pdf(x, mean=mu_gmm[:,i], cov=gmm_cov[i])*Alpha[i]
    return w/w.sum()
for i in range(3):
    W = np.apply_along_axis(calculate_w_new,0,feature_x)
    for i in range(3):
        diff = feature_x-mu_gmm[:,i,None]
        gmm_cov[i] = np.dot(np.multiply(diff,W[i]),diff.transpose())/N[i]
```

```python
T = 0.7

for i in range(100):
    G1 = im_grey>T
    G2 = im_grey<=T

    G1_mean = im_grey[G1].flatten().mean()
    G2_mean = im_grey[G2].flatten().mean()
#     print(T-(G1_mean+G2_mean)/2.0)
    T = (G1_mean+G2_mean)/2.0
```

```python
plt.imshow(im_grey>T)
o = (im_grey>T)*1
im_otsu = (im_grey*255).astype(np.uint8)
# o[o==1] = 255
# PIL.Image.fromarray(o.astype(np.uint8)).save('global_T.bmp')
    
```

### Otsu 

```python
p = plt.hist(im_otsu.flatten(),bins=256,range=(0,256),normed=True)
p = p[0]
```

```python
P1 = np.cumsum(p)
P2 = 1.0-P1
m1 = np.zeros(256)
for i in range(256):
    index = np.array(range(i+1))
    m1[i] = np.multiply(index,p[:i+1]).sum()/P1[i]
    
m2 = np.zeros(256)
for i in range(255):
    index = np.arange(i+1,256)
    m2[i] = np.multiply(index,p[i+1:256]).sum()/P2[i]
```

```python
sigmaB = P1*P2*np.square(m1-m2)
```

```python
K = sigmaB.argmax()
K
```

```python
plt.imshow(im_grey>K/(255.0))
plt.show()
```

```python
print('K:',K,'T',T*255.0)
```

```python
T = 0.1
def global_K(im,T):
    T = 0.1
    for i in range(10):
        G1 = im>T
        G2 = im<=T
    
        G1_mean = im[G1].flatten().mean()
        G2_mean = im[G2].flatten().mean()
    #     print(T-(G1_mean+G2_mean)/2.0)
        T = (G1_mean+G2_mean)/2.0
        return T
```

### 
5*10gridglobal threshold 

```python
im_grid = []
col_grid_width = im_grey.shape[1]//10
row_grid_width = im_grey.shape[0]//5

for i in range(5):
    tem_row = []
    for j in range(10):
        tem = im_grey[i*row_grid_width:(i+1)*row_grid_width,j*col_grid_width:(j+1)*col_grid_width].copy()
        tem_row.append(tem)
    im_grid.append(tem_row)
im_local_ostu = np.zeros(im_grey.shape)
# for i in range(5):
#     for j in range(5):
#         im_local_ostu[i*row_grid_width:(i+1)*row_grid_width,j*col_grid_width:(j+1)*col_grid_width] = ostu((im_grid[i][j]*255.0).astype(np.uint8))*1.0
        
```

```python

f,ax = plt.subplots(5,10)
im_local_ostu = np.zeros(im_grey.shape)
for i in range(5):
    for j in range(10):
        K = global_K((im_grid[i][j]),T)
        im_local_ostu = (im_grey<K)*1.0
        ax[i][j].imshow(im_grid[i][j])

f.show()
```
